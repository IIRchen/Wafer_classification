```python
import numpy as np
data = np.load('/content/drive/MyDrive/DeepLearning_project/Dataset/Wafer_Map_Datasets.npz')
print(data)
img_array = data['arr_0']
label_array = data['arr_1']
```

```python
from tqdm import tqdm
import numpy as np

def check_and_modify(array):
    array[array > 2] = 2
    array[array < 0] = 0
    return array


for index in tqdm(range(len(img_array))) :
  check_image = check_and_modify(img_array[index])
  img_array[index] = check_image
```
