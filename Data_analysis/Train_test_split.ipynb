```python
import numpy as np
data = np.load('/content/drive/MyDrive/DeepLearning_project/Dataset/Wafer_Map_Datasets.npz')
img_array = data['arr_0']
label_array = data['arr_1']
```

```python
from sklearn.model_selection import train_test_split

# 将数据集拆分成训练集和验证集
X_train, X_val, y_train, y_val = train_test_split(img_array, label_array, test_size=0.2, stratify=label_array)

# 如果需要的话，可以查看拆分后的数据集大小
print("训练集大小:", X_train.shape, y_train.shape)
print("验证集大小:", X_val.shape, y_val.shape)
```

```python
np.savez("/content/drive/MyDrive/DeepLearning_project/Dataset/Training_dataset.npz", arr_0=X_train, arr_1=y_train)
```

```python
np.savez("/content/drive/MyDrive/DeepLearning_project/Dataset/Val_dataset.npz", arr_0=X_val, arr_1=y_val)
```
